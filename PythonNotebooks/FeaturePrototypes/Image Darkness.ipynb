```python
import argparse
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

#Image luminosity from http://stackoverflow.com/questions/596216/formula-to-determine-brightness-of-rgb-color
#Image Darkness Measured by Histogram shape (http://docs.opencv.org/trunk/d1/db7/tutorial_py_histogram_begins.html)

import numpy as np
image = cv2.imread('/Users/richardknoche/Desktop/TestImages/Darkness/Dim/Dim5.jpg')
hist = cv2.calcHist([image],[0],None,[256],[0,256])

fig, ax = plt.subplots()
#ax[0].bar(np.arange(0,256,1),hist)
ax.imshow( cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
b,g,r = np.average(np.average(image,axis=0),axis=0)
standard_luminance = (0.2126*r + 0.7152*g + 0.0722*b)
percieved_luminace = (0.299*r + 0.587*g + 0.114*b)
print('Standard Luminance: ',standard_luminance)
print('Percieved Luminance: ',percieved_luminace)
```

```python
hist,bins,_ = plt.hist(image.ravel(),255,[0,255])
bin_centers = (bins[1:]+bins[:-1])/2
(bin_centers*hist).sum()/sum(hist)
```
