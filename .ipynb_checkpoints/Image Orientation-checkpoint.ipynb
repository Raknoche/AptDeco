```python
#Doesn't seem to work very well... hard to distinguish vertical lines from horizontal lines, and background from foreground

import argparse
import cv2
from matplotlib import pyplot as plt
import numpy as np
import pycircstat.descriptive as circstat
%matplotlib inline

image = cv2.imread('/Users/richardknoche/Desktop/TestImages/Blur/Sharp/Couch2.jpg')
scaler = np.min([800.0/image.shape[0], 800.0/image.shape[1]])
image = cv2.resize(image,(np.int(scaler*image.shape[1]),np.int(scaler*image.shape[0])),interpolation=cv2.INTER_AREA)

I_hsv = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
I_h = I_hsv[:,:,0]
I_s = I_hsv[:,:,1]
I_v = I_hsv[:,:,2]
I_g = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)


def CalcLineAngle(src):
    
    lines = cv2.HoughLinesP(cv2.Canny(src,100,300,apertureSize = 3),1,np.pi/180,50,minLineLength = 50, maxLineGap = 5)
    if lines is not None:
        avg_line_angle = (180/np.pi)*circstat.mean([ [np.arctan2(np.abs(y2-y1),np.abs(x2-x1)) for x1,y1,x2,y2 in lines[idx]] for idx in range(len(lines))])
    else:
        avg_line_angle = None

    return avg_line_angle


I_h_angles = CalcLineAngle(I_h)
I_s_angles = CalcLineAngle(I_s)
I_v_angles = CalcLineAngle(I_v)
```
