```python
import argparse
import cv2
from matplotlib import pyplot as plt
import imutils
import numpy as np
%matplotlib inline
import pandas as pd

file = '/Users/richardknoche/Desktop/TestImages/Directionality/Front.jpg'
image = cv2.imread(file)
img = cv2.imread(file,0)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
#blur = cv2.GaussianBlur(img,(5,5),0)
ret3,thresh = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
thresh=cv2.bitwise_not(thresh)
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
```

```python
#Find contours
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cnts = cnts[0] if imutils.is_cv2() else cnts[1]

# loop over the contours
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))

#Also grab other information while we're at it
areas = np.array([]) #fraction of image occupied by object
widths = np.array([]) #item widths (normalized to image width)
heights = np.array([]) #item heights (normalized to image height)
extents = np.array([]) #fraction of rectangular bounding box that is occupied
left_weights = np.array([]) #(x_center - x_min)/width -- how left heavy is it
top_weights = np.array([]) #(y_center - y_min)/height -- how top heavy is it
aspect_ratios = np.array([]) #ratio of width to height
solidities = np.array([]) #How "solid" the object is

for c in cnts:
    # compute the center of the contour
    M = cv2.moments(c)
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    plt.scatter(cX,cY,color='g',s=100,marker='+',linewidth=3)

    #contour properties
    area = cv2.contourArea(c)
    x,y,w,h = cv2.boundingRect(c)
    rect_area = w*h
    hull = cv2.convexHull(c)
    hull_area = cv2.contourArea(hull)

    #Furniture Features
    areas = np.append(areas,float(area))
    extents = np.append(extents,float(area)/rect_area)
    widths = np.append(widths,float(w)/image.shape[1])
    heights =np.append(heights,float(h)/image.shape[0])
    left_weights = np.append(left_weights, (cX - x)/w)
    top_weights = np.append(top_weights, (cY - y)/w)
    aspect_ratios = np.append(aspect_ratios, float(w)/h)
    solidities = np.append(solidities,float(area)/hull_area)

    
#Weight of pixels
df = pd.DataFrame(thresh)
top_count = df.ix[0:np.floor(thresh.shape[0]/2),:].sum().sum()/255
bottom_count = df.ix[np.floor(thresh.shape[0]/2):,:].sum().sum()/255
left_count = df.ix[:,0:np.floor(thresh.shape[1]/2)].sum().sum()/255
right_count = df.ix[:,np.floor(thresh.shape[1]/2):].sum().sum()/255
left_weight_2 = left_count/(left_count+right_count)
top_weight_2 = top_count/(top_count+bottom_count)

#Area of the contours
contour_areas = np.array([area for area in areas])/thresh.size

#Full area of the contours
full_area = areas.sum()/thresh.size

#Draw the contour (assuming there is only one)
xx = [coord[0][0] for coord in c]
yy = [coord[0][1] for coord in c]

plt.plot(xx,yy,color='g',linewidth=3)

#Image dimensions
half_xl = np.floor(thresh.shape[1]/2)
half_yl = np.floor(thresh.shape[0]/2)

#Text info
print('Number of Pieces: %d' % (len(cnts)))
print('--- Centroid Features ---')
print('X Offset %0.2f' % ((cX-half_xl)/(half_xl*2)) )
print('Y Offset %0.2f' % ((cY-half_yl)/(half_yl*2)) )
print('Center-to-Center Distance: %0.2f' % ( ((cY-half_yl)/(half_yl*2))**2 + ((cX-half_xl)/(half_xl*2))**2)**(1/2) )
print(' ')


print('--- Area Features ---')
print('Fraction of Image: %0.2f' % (full_area))
print('Fraction of Bounding Box: %0.2f' % (extents.mean()))
print('Solidity: %0.2f' % (solidities.mean()))
print('Average Contour Width: %0.2f' % (widths.mean()))
print('Average Contour Height: %0.2f' % (heights.mean()))
print(' ')


print('--- Perspective Features ---')
print('Left Weight: %0.2f' % (left_weights.mean()))
print('Top Weight: %0.2f' % (top_weights.mean()))
print('Left Weight v2: %0.2f' % (left_weight_2))
print('Top Weight v2: %0.2f' % (top_weight_2))
print('Aspect Ratio: %0.2f' % (aspect_ratios.mean()))
```
