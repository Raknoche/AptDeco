# Image Symmetry

```python
#From https://pdfs.semanticscholar.org/42b5/33621f2defe97cb6894eefbb6126ce1f2691.pdf
#Example
image=np.zeros([800,800])
image[200:300][:]=255
image[550:600][:]=255
image[550:600][:]=115
image[:,200:300]=115
image[:,550:600]=115
tmp=image

a = tmp.astype("float")
b1 = tmp[::-1,:].astype("float") #Flip upsidedown
b2 = tmp[:,::-1].astype("float") #Flip left/right

#Calculate symmetry by multplying mirrored images
fs = (a + b2)/2
fa = (a - b2)/2
Sym_Horizontal_Value = (fs**2).sum()/((fs**2).sum() + (fa**2).sum())

fs = (a + b1)/2
fa = (a - b1)/2
Sym_Vertical_Value = (fs**2).sum()/( (fs**2).sum() + (fa**2).sum())

#Instead of 0.5-1, scale from 0-1
Sym_Horizontal_Value=Sym_Horizontal_Value*2 - 1
Sym_Vertical_Value=Sym_Vertical_Value*2 - 1

plt.imshow(tmp)
print('Horizontal Value Symmetry: ',Sym_Horizontal_Value)
print('Vertical Value Symmetry: ',Sym_Vertical_Value)
```

```python
#From https://pdfs.semanticscholar.org/42b5/33621f2defe97cb6894eefbb6126ce1f2691.pdf
def CalcSymmetry(src):
    a = src.astype("float")
    b1 = src[::-1,:].astype("float") #Flip upsidedown
    b2 = src[:,::-1].astype("float") #Flip left/right

    #Calculate symmetry by multplying mirrored images
    fs = (a + b2)/2
    fa = (a - b2)/2
    Sym_Horizontal_Value = (fs**2).sum()/((fs**2).sum() + (fa**2).sum())

    fs = (a + b1)/2
    fa = (a - b1)/2
    Sym_Vertical_Value = (fs**2).sum()/( (fs**2).sum() + (fa**2).sum())

    #Instead of 0.5-1, scale from 0-1
    Sym_Horizontal_Value=Sym_Horizontal_Value*2 - 1
    Sym_Vertical_Value=Sym_Vertical_Value*2 - 1

    return(Sym_Horizontal_Value,Sym_Vertical_Value)


image = cv2.imread('/Users/richardknoche/Desktop/TestImages/Blur/Sharp/Couch2.jpg')
I_hsv = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
I_h = I_hsv[:,:,0]
I_s = I_hsv[:,:,1]
I_v = I_hsv[:,:,2]


Sym_Horizontal_Hue,Sym_Vertical_Hue = CalcSymmetry(I_h)
Sym_Horizontal_Saturation,Sym_Vertical_Saturation = CalcSymmetry(I_s)
Sym_Horizontal_Value,Sym_Vertical_Value = CalcSymmetry(I_v)


plt.imshow(image)
print('Horizontal Value Symmetry: ',Sym_Horizontal_Value)
print('Vertical Value Symmetry: ',Sym_Vertical_Value)

print('Horizontal Saturation Symmetry: ',Sym_Horizontal_Saturation)
print('Vertical Saturation Symmetry: ',Sym_Vertical_Saturation)
```

```python
import argparse
import cv2
from matplotlib import pyplot as plt
import imutils
import numpy as np
%matplotlib inline
import pandas as pd

file='/Users/richardknoche/Desktop/TestImages/Blur/Sharp/Couch6.jpg'
image = cv2.imread(file)
img = cv2.imread(file,0)
img = cv2.GaussianBlur(img,(5,5),30)
ret3,thresh = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
thresh=cv2.bitwise_not(thresh)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if imutils.is_cv2() else cnts[1]

#Also grab other information while we're at it
cXs=np.array([])
cYs=np.array([])

for c in cnts:
    # compute the center of the contour
    M = cv2.moments(c)
    if  M["m00"] != 0:
        cXs = np.append(cXs,int(M["m10"] / M["m00"]))
        cYs = np.append(cYs,int(M["m01"] / M["m00"]))
        #plt.scatter(int(M["m10"] / M["m00"]),int(M["m01"] / M["m00"]),color='g',s=100,marker='+',linewidth=3)

busyness = ( (cXs.std()/cXs.mean())**2 + (cYs.std()/cYs.mean())**2)**(1/2)  
print(len(cnts))
print(busyness)
```
